## Masa de un sólido
La consigna es la siguiente, tenés que encontrar mediante el método de montecarlo la masa de un tetraedro delimitado por el plano $x + 2y + 3z = 6$ y el primer octante, como se vé en la figura.<br>
<img src="./fig_11_7_tetrahedron (1).svg"><br>
La densidad de cada punto $(x,y,z)$ esta dado por $\rho(x,y,z) = x\,y^2\,z$ <br><br>
Si quisieras resolverlo analíticamente, la masa del tetraedro se puede encontrar mediante la integral triple<br>
\begin{equation}
M = \iiint_{S}\rho(x,y,z) dV
\end{equation}
<br>
Donde $S$ es el sólido descripto arriba. <br> 
Para encontrar los límites de integración, despejemos   $z$ del plano inclinado  y nos queda $ 0 < z < \frac{1}{3}(6 - x - 2y)$, para encontrar $x$ e $y$ proyectamos el tetraedro en el plano $xy$ correspondiente a $z = 0$ resultando en $x + 2y = 6$, nos queda $ 0 < y < 3 - \frac{1}{2}x$ ,  finalmente, $ 0 < x < 6$ <br>
Nuestra integral analítica entonces es la siguiente<br>
\begin{equation}
M = \int_6^0 \int_{0}^{3-\frac{1}{2}x} \int_{0}^{\frac{1}{3}(6-x-2y)} \rho(x,y,z) dz dy dx
\end{equation}
<br>
Para resolver este ejercicio por el método de Montecarlo, debés crear los puntos en 3 coordenadas con números aleatorios y evaluar que caigan en el volumen delimitado por las paredes del cubo y el plano.

```python
def densidad(x,y,z):
    return x* y**2 * z
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')

i=0
n = 10000
m = 0
lim_x = 6
lim_y = 3
lim_z = 2
vol = lim_x * lim_y * lim_z

while  i < n:
     #segun mis datos, mis nros random van todos de cero hasta donde corta el plano sobre el eje.
     rx = np.random.random()* lim_x
     ry = np.random.random()* lim_y
     rz = np.random.random()* lim_z

     if  6 >= rx + 2*ry + 3*rz  : # solo la voy a tener en cuenta si el modulo de la posición es menor al radio de la esfera (es decir, la carga está dentro de ella)
          ax.scatter(rx,ry,rz, alpha=0.2)
          m += densidad(rx,ry,rz)
     i +=1

plt.show()
print(f'la masa del cuerpo es {m * vol  / n}')
```
